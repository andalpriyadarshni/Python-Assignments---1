```python
def upperlower(l):
      upper = 0
      lower = 0
      for i in l:
          if i.isupper() :
              upper = upper+1
          elif i.islower():
              lower = lower +1


      print("THE UPPERCASE APPEARS", upper ,"times in the STRING")
      print("THE LOWERCASE APPEARS", lower ,"times in the STRING")


upperlower('HELlowoRLD')
```

```python
#Vowel Counter
Write a program to count how many vowels are in a given string (ignore case).
```

```python
def vow():
      x = input("Enter the string: ")
      vowels = 'aeiouAEIOU'
      vowels_found = []
      for char in x :
         if char in vowels :
             vowels_found.extend(char)
      print('FOUND the vowels:', vowels_found)

vow()



```

```python
#Common Elements in Two Lists
Write a function to return a list of elements common to both lists.
➡ Example: [1,2,3,4] and [3,4,5,6] → [3,4]
```

```python
#Reverse Words in a Sentence
Write a function that reverses the order of words in a given sentence.
➡ Example: "I love Python" → "Python love I"
```

```python
def rev():
  x = input("Enter the string: ")
  result = x[::-1]
  print("reverse string: ",result)

rev()
```

```python
def rev():
  x = input("Enter the string: ")
  words = x.split()
  result = " ".join(words[::-1])
  print("reverse string: ",result)

rev()
```

```python
#Sum of Digits
Take an integer input and return the sum of its digits.
➡ Example: 1234 → 10
```

```python
def find_sum():
    x = input("Enter the numbers: ")
    numbers = list(map(int, x.split()))
    result = sum(numbers)
    print("Sum of the numbers:", result)


find_sum()
```

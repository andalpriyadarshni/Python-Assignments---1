```python
x = float(input('Enter your score (0 to 10) : ' ))
if x > 7 :
     print(f"{x} is Above Average,GOOD PERFORMANCE")
elif 4 <= x <= 7 :
     print(f"{x} is Average,NEEDS TO IMPROVE")
else :
     print(f"{x} is Below Average,CONTINUOUS PRACTISE MUST NEEDED TO GET GOOD RESULT")
```
